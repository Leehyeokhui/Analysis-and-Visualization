```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2015', periods=1000))
s.plot()
```

```python
s = s.cumsum() # 누적합
s.plot()
```

```python
r = s.rolling(window=60)
r.mean()
# 이동평균
```

```python
s.plot(style='k--')
r.mean().plot(style='k')
```

```python
r = s.rolling(window=30)
s.plot(style='k--')
r.mean().plot(style='k')
```

```python
df = pd.DataFrame(np.random.randn(1000,4), index=pd.date_range('1/1/2015', periods=1000),
                 columns=['A', 'B', 'C', 'D'])
df.head()
```

```python
df = df.cumsum()
df
```

```python
df.plot()
df.rolling(window=60).sum().plot(subplots=True)
```

```python
df.rolling(window=len(df), min_periods=1).mean().plot()
```

```python
df.expanding(min_periods=1).mean().plot()
```
